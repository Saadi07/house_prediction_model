```python
DFS/BFS
def dfs(graph, node, visited):
    if node not in visited:
        visited.append(node)
        for k in graph[node]:
            dfs(graph,k, visited)
    return visited

graph = {'A':['B','C','D'], 
         'B': ['G','H'], 
         'C': [],
         'D': ['E','F'], 
         'E': ['I','J'], 
         'F': [], 
         'G': [],
         'H': [], 
         'I': [],
         'J': []
        }

visited = dfs(graph,'A', [])
print(visited)
```
